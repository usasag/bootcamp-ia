```python
import requests
import zipfile
import os

url = "https://download.microsoft.com/download/3/e/1/3e1c3f21-ecdb-4869-8368-6deba77b919f/kagglecatsanddogs_5340.zip"
output_filename = "cats_vs_dogs.zip"
extract_folder = "cats_vs_dogs_dataset"

# Download the file
print(f"Downloading {url}...")
response = requests.get(url, stream=True)
response.raise_for_status() # Raise an exception for bad status codes

with open(output_filename, 'wb') as file:
    for chunk in response.iter_content(chunk_size=8192):
        file.write(chunk)
print(f"Downloaded {output_filename}")

# Extract the zip file
print(f"Extracting {output_filename} to {extract_folder}...")
with zipfile.ZipFile(output_filename, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)
print("Extraction complete.")

# Clean up the downloaded zip file
# os.remove(output_filename)
# print(f"Removed {output_filename}")

# You can now access the dataset in the 'cats_vs_dogs_dataset' folder.
# The actual images are likely nested within subfolders inside 'datasets'.
# You might need to navigate through the extracted structure to find the image files.
print(f"Dataset extracted to: {extract_folder}")
# List contents of the extracted folder to see the structure
!ls -R {extract_folder}
```

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import keras
from tensorflow.keras.applications import VGG19
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import logging
import os

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Define dataset path
dataset_path = './cats_vs_dogs_dataset/PetImages/'
train_dir = os.path.join(dataset_path, 'Train') # Create train and validation directories
validation_dir = os.path.join(dataset_path, 'Validation')

# Create directories for train and validation if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(validation_dir, exist_ok=True)

# Organize data into train and validation sets
def organize_data(source_dir, train_dir, validation_dir, split_ratio=0.8):
    classes = ['Cat', 'Dog']
    for cls in classes:
        class_source_dir = os.path.join(source_dir, cls)
        class_train_dir = os.path.join(train_dir, cls)
        class_validation_dir = os.path.join(validation_dir, cls)

        os.makedirs(class_train_dir, exist_ok=True)
        os.makedirs(class_validation_dir, exist_ok=True)

        all_files = [os.path.join(class_source_dir, f) for f in os.listdir(class_source_dir) if os.path.isfile(os.path.join(class_source_dir, f))]
        np.random.shuffle(all_files)
        train_files_count = int(len(all_files) * split_ratio)

        for i, file_path in enumerate(all_files):
            try:
                if i < train_files_count:
                    destination_dir = class_train_dir
                else:
                    destination_dir = class_validation_dir
                # Check if the file is valid before moving
                img = tf.keras.utils.load_img(file_path) # Use tf.keras.utils.load_img
                img.close() # Close the PIL image object
                os.rename(file_path, os.path.join(destination_dir, os.path.basename(file_path)))
            except Exception as e:
                logging.warning(f"Skipping invalid file {file_path}: {e}")

logging.info("Organizing data into train and validation sets...")
organize_data(dataset_path, train_dir, validation_dir)
logging.info("Data organization complete.")

# Define image dimensions and batch size
img_width, img_height = 224, 224
batch_size = 64 # Changed batch size to 64
epochs = 30 # Changed epochs to 30

# Data Augmentation and Preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

# Load the VGG19 model with pre-trained ImageNet weights
logging.info("Loading VGG19 model with ImageNet weights...")
base_model = VGG19(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))

# Freeze the convolutional base
for layer in base_model.layers:
    layer.trainable = False

# Add custom classification layers
x = Flatten()(base_model.output)
x = Dense(256, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x) # Use sigmoid for binary classification

model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model
logging.info("Compiling the model...")
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# Train the model
logging.info(f"Starting training for {epochs} epochs...")
# Use tf.distribute.MirroredStrategy for CUDA acceleration if multiple GPUs are available
strategy = tf.distribute.MirroredStrategy()
with strategy.scope():
    # Re-create the model inside the strategy scope for distributed training
    base_model = VGG19(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))
    for layer in base_model.layers:
        layer.trainable = False
    x = Flatten()(base_model.output)
    x = Dense(256, activation='relu')(x)
    predictions = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // batch_size,
        epochs=epochs,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // batch_size)

logging.info("Training complete.")

# Plot training and validation loss
logging.info("Plotting training and validation loss...")
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.show()

logging.info("Script finished.")
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg19 import preprocess_input
import requests
import random
import os

# substitua essas URLs por qualquer imagem de gato ou cachorro correspondente.
cat_urls = [
    "https://jpimg.com.br/uploads/2025/01/10-curiosidades-sobre-os-filhotes-de-gato.jpg"
]

dog_urls = [
    "https://images.unsplash.com/photo-1561648078-5b5f0a645d7b?ixlib=rb-1.2.1&auto=format&fit=crop&w=1350&q=80",
    "https://images.unsplash.com/photo-1598133894008-61f7fdb8cc3a?ixlib=rb-1.2.1&auto=format&fit=crop&w=1350&q=80"
]

# Função para baixar uma imagem de uma URL
def download_image(url, filename):
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()
        with open(filename, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        print(f"Imagem baixada: {filename}")
        return True
    except Exception as e:
        print(f"Erro ao baixar a imagem de {url}: {e}")
        return False

# Baixar uma imagem aleatória de gato
cat_image_url = random.choice(cat_urls)
cat_filename = 'gato.jpg'
if download_image(cat_image_url, cat_filename):
    # Carregar e pré-processar a imagem do gato
    try:
        img = image.load_img(cat_filename, target_size=(img_width, img_height))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = preprocess_input(img_array)

        # Fazer a previsão para o gato
        prediction = loaded_model.predict(img_array)

        if prediction[0] < 0.5:
            print(f"'{cat_filename}' é provavelmente um Gato. Pontuação de confiança: {1 - prediction[0][0]:.4f}")
        else:
            print(f"'{cat_filename}' é provavelmente um Cachorro. Pontuação de confiança: {prediction[0][0]:.4f}")

        print(f"Valor de previsão bruto para '{cat_filename}': {prediction[0][0]:.4f}")
    except Exception as e:
        print(f"Erro ao processar a imagem '{cat_filename}': {e}")
    finally:
        # Limpar o arquivo baixado
        if os.path.exists(cat_filename):
            os.remove(cat_filename)


# Baixar uma imagem aleatória de cachorro
dog_image_url = random.choice(dog_urls)
dog_filename = 'cachorro.jpg'
if download_image(dog_image_url, dog_filename):
    # Carregar e pré-processar a imagem do cachorro
    try:
        img = image.load_img(dog_filename, target_size=(img_width, img_height))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = preprocess_input(img_array)

        # Fazer a previsão para o cachorro
        prediction = loaded_model.predict(img_array)

        if prediction[0] < 0.5:
            print(f"'{dog_filename}' é provavelmente um Gato. Pontuação de confiança: {1 - prediction[0][0]:.4f}")
        else:
            print(f"'{dog_filename}' é provavelmente um Cachorro. Pontuação de confiança: {prediction[0][0]:.4f}")

        print(f"Valor de previsão bruto para '{dog_filename}': {prediction[0][0]:.4f}")
    except Exception as e:
        print(f"Erro ao processar a imagem '{dog_filename}': {e}")
    finally:
        # Limpar o arquivo baixado
        if os.path.exists(dog_filename):
            os.remove(dog_filename)
```
